```python
import pandas as pd


data = pd.read_csv('../../../data/parks.csv')
```

```python
print(data['Location'].unique())
```

```python
import matplotlib.pyplot as plt

#plt.figure(figsize=(30,30))
plt.bar(data['Location'].unique(),data.groupby('Location')['Park'].count())
plt.title('No of Parks per location', fontsize=20)
plt.xlabel('Location', fontsize=15)
plt.ylabel('Count of Parks', fontsize=15)
plt.xticks(rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt


df = pd.read_csv('../../../data/parks.csv')
df['features'] = df['Feature(s)'].str.split(',')

df = df.explode('features')
park_counts = df.groupby('features').size()
park_counts.plot(kind='bar')

# Set the plot title and axis labels
plt.title('Number of Parks per Feature', fontsize=20)
plt.xlabel('Feature', fontsize=15)
plt.ylabel('Number of Parks', fontsize=15)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../../data/parks.csv')
df['features'] = df['Feature(s)'].str.split(',')
df = df.explode('features')

# Filter the dataframe by the Location column to get the parks in South Boston
d4Dor_df = df[df['Location'] == 'Dorchester'] 

#or (df['Location'] == 'Mattapan') or (df['Location'] == 'Roslindale') or (df['Location'] == 'Jamaica Plain')
# Group by the features column and count the number of parks
park_counts = d4Dor_df.groupby('features').size()

# Plot the bar chart
park_counts.plot(kind='bar')

# Set the plot title and axis labels
plt.title('Number of Parks per Feature in Dorchester', fontsize=20)
plt.xlabel('Feature', fontsize=15)
plt.ylabel('Number of Parks', fontsize=15)

# Display the plot
plt.show()
```
