```python
import numpy as np
import matplotlib as plt
import pandas as pd

shootingdf = pd.read_csv("Shootings.csv")
df1 = shootingdf.copy()
dist4 = '|'.join(['B2','B3','C11','E13'])
other = '|'.join(['A15','A1','A7','C6','D4','D14','E5','E18'])
df1['district'] = df1['district'].str.replace(dist4, 'District 4')
df1['district'] = df1['district'].str.replace(other, 'Not District 4')

df1
```

```python
probabilities = df1['district'].value_counts(normalize=True)
print('Percentages of Shootings in District 4:' , probabilities[0]*100)
print('Percentages of Shootings in All Other Districts:', probabilities[1]*100)
total = (probabilities[0] + probabilities[1])*100
print('Summation of the Two:', total)
probabilities.plot(kind='bar', figsize=(10,5), color='blue', title='Probabilities of Shootings in District 4 vs. All Other Districts')
```
