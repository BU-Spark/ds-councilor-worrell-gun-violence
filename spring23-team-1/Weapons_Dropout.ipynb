```python
publicDF = pd.read_csv("data/publicSchools.csv")
publicDF['Address 1'] = publicDF['Address 1'].astype(str)
publicDF['Town'] = publicDF['Town'].astype(str)
publicDF['State'] = publicDF['State'].astype(str)
publicDF['Zip'] = publicDF['Zip'].astype(str)
publicDF['Address'] = publicDF[['Address 1', 'Town', 'State', 'Zip']].agg(','.join, axis=1)
publicDF.head(6)
```
