```python
import pandas as pd
import numpy as np
```

Files that used:

```python
shapefile_df = pd.read_csv('shapefile.csv') # compare the difference between different district
shooting_df = pd.read_csv('shootings.csv')
```

```python
temp = shooting_df.groupby('district').agg({'district': 'count'}).rename(columns={'district': 'shooting_count'}).reset_index()
```

Belowed is the shooting number in each district

```python
temp.plot.bar(x='district', y='shooting_count',rot = 45)
```

```python
temp = pd.DataFrame()
temp['district'] = shooting_df['district'].unique()
temp['Fatal']= shooting_df[shooting_df['shooting_type_v2'] == 'Fatal'].groupby(['district']).agg({'district': 'count'}).rename(columns={'district': 'x'}).reset_index()['x']
temp['non_Fatal']= shooting_df[shooting_df['shooting_type_v2'] == 'Non-Fatal'].groupby(['district']).agg({'district': 'count'}).rename(columns={'district': 'x'}).reset_index()['x']
#shooting_df.groupby(['district','shooting_type_v2'])['district'].agg('count').reset_index(name='counts')
```

We compare the fatal shooting number and non fatal shooting number in each district. Found that C11, E18, E5, A7, A15 generally has lower shooting cases. We intend to find the different between these districts and the other district with higher shooting number.
